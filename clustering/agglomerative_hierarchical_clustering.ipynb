```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# Step 1: Generate synthetic data
X, y = make_blobs(n_samples=10, centers=3, random_state=42)

# Step 2: Perform Agglomerative Hierarchical Clustering
# 'ward' minimizes the variance within each cluster being merged
Z = linkage(X, method='ward')

# Step 3: Plot the Dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Agglomerative Hierarchical Clustering Dendrogram')
plt.xlabel('Data points')
plt.ylabel('Distance')
plt.show()

# Step 4: Form flat clusters by cutting the dendrogram at a specific distance
max_d = 7.0  # Max distance to cut the dendrogram
clusters = fcluster(Z, max_d, criterion='distance')

# Step 5: Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o')
plt.title('Agglomerative Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
