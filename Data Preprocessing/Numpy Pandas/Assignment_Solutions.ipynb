Broadcasting is a mechanism that allows NumPy to perform element-wise operations on arrays of different shapes by expanding the smaller array to match the shape of the larger array.

In this case, a has a shape of (2, 3) and b has a shape of (3,).
NumPy will broadcast b to match the shape of a. It will replicate b along the first axis (rows) to match a's shape.
So the array b will be transformed and then added to a resulting in the given matrix.

```python
# Question 3: 
# Create a 1D array of random numbers and find the indices of local peaks (values that are strictly greater than their neighbors).
# (A local peak is an element in an array that is greater than both its left and right neighbors.)
# (Hint : create a random 1D array, and use np.where() function to put a condition, relate it to slicing and use comparisons to formulate 
#  the function.)
import numpy as np
array = np.random.rand(20)  # 20 random floats between 0 and 1
print(array)
peaks = np.where((array[1:-1] > array[:-2]) & (array[1:-1] > array[2:]))[0] + 1
print(peaks)


```

DRY RUN<br>
Suppose array is: [3, 5, 1, 4, 6, 3, 2].<br>
FIRST CONDITION:<br>
<br>
array[1:-1]: This slices the array and gets all elements except the first and last.<br>
For example, if array is [3, 5, 1, 4, 6, 3, 2], then array[1:-1] will be [5, 1, 4, 6, 3].<br>
array[:-2]: This gets all elements except the last two.<br>
For example, [3, 5, 1, 4, 6, 3, 2] becomes [3, 5, 1, 4, 6].<br>
array[2:]: This gets all elements except the first two.<br>
For example, [3, 5, 1, 4, 6, 3, 2] becomes [1, 4, 6, 3, 2].<br>
<br>
SECOND CONDITION:<br>
(array[1:-1] > array[:-2]): This checks if each element (excluding the first and last) is greater than the element before it.<br>
In our example: [5 > 3, 1 > 5, 4 > 1, 6 > 4, 3 > 6] → [True, False, True, True, False].<br>
(array[1:-1] > array[2:]): This checks if each element (excluding the first and last) is greater than the element after it.<br>
In our example: [5 > 1, 1 > 4, 4 > 6, 6 > 3, 3 > 2] → [True, False, False, True, True].<br>
finally, using np.where() will give us the peaks, that is when BOTH the conditions are true.<br>

Answer the following: 
<br>
a)	How can you identify any outliers in the Age and Grade columns using pandas?<br>
b)	How can you group the students based on City and calculate the average Grade for each city?<br>
c)	 How can you filter out the students who have a Grade greater than 80 and are from the city of "New York"?<br>
d)	How can you sort the dataset based on Grade in descending order and then by Age in ascending order, if two students have the same grade?<br>
e)	 How can you create a new column Pass/Fail, which categorizes students as 'Pass' if their Grade is 80 or above and 'Fail' otherwise?<br>


```python
import pandas as pd
df = pd.read_csv("C:\\Users\\DELL\\OneDrive\\Desktop\\Untitled spreadsheet.csv")
#a)
# Finding Age Outliers using IQR
Q1_age = df['Age'].quantile(0.25)
Q3_age = df['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age

age_outliers = df[(df['Age'] < (Q1_age - 1.5 * IQR_age)) | (df['Age'] > (Q3_age + 1.5 * IQR_age))]
print("Age Outliers:\n", age_outliers)

# Finding Grade Outliers using IQR
Q1_grade = df['Grade'].quantile(0.25)
Q3_grade = df['Grade'].quantile(0.75)
IQR_grade = Q3_grade - Q1_grade

grade_outliers = df[(df['Grade'] < (Q1_grade - 1.5 * IQR_grade)) | (df['Grade'] > (Q3_grade + 1.5 * IQR_grade))]
print("Grade Outliers:\n", grade_outliers)




```

```python
#b)
# Grouping by City and calculating mean Grade
city_group = df.groupby('City')['Grade'].mean()
print(city_group)

```

```python

#c) # Filter students with Grade > 80 and from New York
filtered_students = df[(df['Grade'] > 80) & (df['City'] == 'New York')]
print(filtered_students)
```

```python
#d)
# # Sort by Grade in descending order, then by Age in ascending order
sorted_df = df.sort_values(by=['Grade', 'Age'], ascending=[False, True])
print(sorted_df)
```

```python
#e)
# Define a custom function
def pass_or_fail(grade):
    if grade >= 80:
        return 'Pass'
    else:
        return 'Fail'

# Apply the custom function to create a new column
df['Pass/Fail'] = df['Grade'].apply(pass_or_fail)
print(df)
```
