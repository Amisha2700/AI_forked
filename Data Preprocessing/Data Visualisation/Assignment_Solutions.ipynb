```python
#QUESTION-1: Plot a histogram to analyze the distribution of total_profit across all months.
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(r"C:\Users\DELL\OneDrive\Desktop\sales.csv")
total_profit = data['total_profit']
plt.figure(figsize=(10, 6))
# Plot a histogram to show the distribution of 'total_profit' values
# 'bins=8' specifies the number of bins to categorize the data into
# 'color' sets the bar color, and 'edgecolor' adds borders to each bar for clarity
plt.hist(total_profit, bins=8, color='lightgreen', edgecolor='black')
plt.title('Distribution of Total Profit Across All Months')
plt.xlabel('Total Profit') #to label x axis
plt.ylabel('Frequency') #to label y-axis
plt.grid(True)
plt.show()
```

```python
# QUESTION-2:Create a box plot to show the distribution of monthly sales (total_units) for each
# product, highlighting outliers.
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\DELL\OneDrive\Desktop\DSML_Labs\A1_Numpy\sales (1).csv")
data = {
    'Facecream': df['facecream'],
    'Facewash': df['facewash'],
    'Toothpaste': df['toothpaste'],
    'Bathingsoap': df['bathingsoap'],
    'Shampoo': df['shampoo'],
    'Moisturizer': df['moisturizer']
}
boxplot_df = pd.DataFrame(data)
plt.figure(figsize=(12, 8))
# Plot a box plot to show the distribution of total units sold by each product
# 'patch_artist=True' fills the boxes with color
# 'showfliers=True' includes outliers in the plot to highlight unusual values
# 'boxprops', 'whiskerprops', 'capprops', and 'medianprops' are used to style the plot
plt.boxplot(boxplot_df, patch_artist=True, showfliers=True, 
            boxprops=dict(facecolor='yellow', color='blue'),    # Color settings for box edges
            whiskerprops=dict(color='blue'),                    # Color of whiskers
            capprops=dict(color='blue'),                        # Color of the caps at ends of whiskers
            medianprops=dict(color='red'))                      # Color of the median line in each box

# Label the x-axis with product names
# 'range(1, len(data) + 1)' generates positions for each product on the x-axis
# 'boxplot_df.columns' holds the product names (Facecream, Facewash, etc.)
# 'rotation=45' rotates the labels for better readability if they are long
plt.xticks(range(1, len(data) + 1), boxplot_df.columns, rotation=45)

# Add labels and title for the plot
plt.xlabel('Product')              # Label for the x-axis
plt.ylabel('Total Units')          # Label for the y-axis
plt.title('Distribution of Monthly Sales (Total Units) by Product')  # Title of the plot

# Adjust the layout so all elements fit neatly
plt.tight_layout()

# Display the plot
plt.show()
```

```python
#QUESTION-3 : Generate a bar chart showing the monthly sales figures of all products (facecream,
# facewash, toothpaste, bathingsoap, shampoo, moisturizer).
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the sales data from the CSV file into a DataFrame
data = pd.read_csv(r"C:\Users\DELL\OneDrive\Desktop\sales.csv")

# Define the list of products to analyze
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

# Extract the 'month_number' column for the x-axis labels
months = data['month_number']

# Determine the number of products to set up the spacing in the bar plot
num_products = len(products)

# Define the width of each bar
bar_width = 0.12

# Create an array of positions for each month on the x-axis
r = np.arange(len(months))

# Set up the figure size for better visualization
plt.figure(figsize=(14, 8))

# Loop through each product and plot its monthly sales as a bar plot
for i, product in enumerate(products):
    # Plot each product's sales data; each product is offset horizontally by `i * bar_width`
    plt.bar(r + i * bar_width, data[product], width=bar_width, label=product)

plt.title('Monthly Sales Figures for All Products')
plt.xlabel('Month Number')
plt.ylabel('Sales (Units)')

# Set custom x-ticks to align with the center of grouped bars for each month
# `xticks` positions are offset by `(bar_width * (num_products - 1) / 2)` for centering
plt.xticks(r + bar_width * (num_products - 1) / 2, months)

# Add a legend (also called key) to distinguish different products, with a title "Products"
plt.legend(title="Products")
plt.show()
```

```python
# QUESTION-4 : Generate a stacked bar chart showing the monthly sales figures of all products
# (facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer).
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(r"C:\Users\DELL\OneDrive\Desktop\sales.csv")
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
months = data['month_number']
# Set up the figure size for better visualization
plt.figure(figsize=(14, 8))
# Initialize a 'bottom' array with zeros for stacking bars for each month
bottom = np.zeros(len(months))

# Loop through each product to add its sales data to the stacked bar chart
for product in products:
    # Plot bar for current product, stacking on top of previous products using 'bottom'
    plt.bar(months, data[product], bottom=bottom, label=product)
    
    # Update 'bottom' to include the current product's values, so the next product stacks on top
    bottom += data[product]
plt.title('Monthly Sales Figures for All Products')
plt.xlabel('Month Number')
plt.ylabel('Sales (Units)')
plt.legend(title="Products")
plt.show()
```

```python
# QUESTION -5:Create a pie chart to depict the proportion of total sales (total_units) contributed by
# each product category over the entire period.
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(r"C:\Users\DELL\OneDrive\Desktop\sales.csv")
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
total_sales = data[products].sum()
# Set up the figure size to make the pie chart readable
plt.figure(figsize=(7, 7))

# Create a pie chart to show the percentage of total sales for each product
# 'autopct' formats each slice with a percentage label, 'startangle' rotates the chart for better visual orientation
plt.pie(total_sales, labels=products, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Total Sales (Total Units) by Product Category')
plt.show()
```

```python
# QUESTION - 6: Create a doughnut chart to depict the proportion of total sales (total_units)
# contributed by each product category over the entire period.
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\DELL\OneDrive\Desktop\sales.csv")
total_sales = df[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].sum()
# first Create a figure and a set of subplots for our doughnut chart
# 'aspect="equal"' makes sure the pie looks like a circle
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(aspect="equal"))

# Create the pie chart but with a 'wedgeprops' parameter to make it a doughnut
# 'width=0.6' makes the pie a doughnut by setting the inner radius
wedges, texts, autotexts = ax.pie(total_sales, labels=total_sales.index, 
                                   autopct='%1.2f%%', startangle=120, 
                                   wedgeprops=dict(width=0.6))

# Set the title for our chart to explain what it shows
plt.title('Proportion of Total Sales by Product Category')

# Customize the appearance of the percentage texts inside the doughnut
plt.setp(autotexts, size=10, weight="bold")  # Make the percentage text bold and a bit bigger
plt.setp(texts, size=12)  # Make the category names a little bigger for clarity
plt.show()






```

```python
# QUESTION - 7 :Generate a grouped bar chart to compare the monthly sales of facecream, facewash,
# and toothpaste for each month.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv(r"C:\Users\DELL\OneDrive\Desktop\sales.csv")
# List the products we want to compare
products = ['facecream', 'facewash', 'toothpaste']

# Get the months for the x-axis
months = data['month_number']

# Get the number of products we're comparing
num_products = len(products)

# Set the width for each bar in the grouped bar chart
bar_width = 0.12
# Create an array of positions for the bars on the x-axis
r = np.arange(len(months))
# Set up the figure size to make it look nice and readable
plt.figure(figsize=(14, 8))

# Loop through each product and create a bar for it
for i, product in enumerate(products):
    # Plot the bars with the right offsets to group them together
    plt.bar(r + i * bar_width, data[product], width=bar_width, label=product)

plt.title('Monthly Sales Figures for All Products')
plt.xlabel('Month Number')
plt.ylabel('Sales (Units)')
plt.xticks(r + bar_width * (num_products - 1) / 2, months)
plt.legend(title="Products")
plt.show()
```

```python
# QUESTION - 8:Create a bubble chart to represent monthly total_units where bubble size indicate
# total_profit.
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\DELL\OneDrive\Desktop\sales.csv")
# Extract the relevant columns for plotting
months = df['month_number']  # Month numbers for the x-axis
total_units = df['total_units']  # Total units sold for the y-axis
total_profit = df['total_profit']  # Total profit to determine bubble size and color

# Set up the figure size to ensure the plot is clear and visible
plt.figure(figsize=(35, 10))
# Create a scatter plot (bubble chart)
plt.scatter(
    months,  # X-axis data: months
    total_units,  # Y-axis data: total units sold
    s=total_profit / 10,  # Size of bubbles based on total profit (scaled down for visibility)
    alpha=0.5,  # Set the transparency of the bubbles
    c=total_profit,  # Color of bubbles based on total profit
    cmap='viridis',  # Colormap for the bubbles (viridis for a nice gradient)
    edgecolors='w',  # White edges around bubbles for better visibility
    linewidth=0.5  # Width of the edge lines around bubbles
)

plt.xlabel('Month Number')
plt.ylabel('Total Units')
plt.title('Monthly Total Units with Bubble Size Representing Total Profit')
plt.colorbar(label='Total Profit') #to add color bar to represent intensities of color for each number
plt.show()
```

```python
# QUESTION - 9 : Create a Gantt chart to visualize a hypothetical timeline for sales campaigns of
# different products based on month_number.
import pandas as pd
import matplotlib.pyplot as plt
campaigns = {
    'Product': ['Facecream', 'Facewash', 'Toothpaste', 'Shampoo', 'Bodywash'],
    'StartMonth': [1, 3, 5, 7, 9],
    'EndMonth': [3, 6, 8, 10, 12]
}
df = pd.DataFrame(campaigns)
# Calculate the duration of each campaign (number of months)
df['Duration'] = df['EndMonth'] - df['StartMonth'] + 1  # +1 to include both start and end months

# Set up the figure size for the Gantt chart
plt.figure(figsize=(14, 8))

# Loop through each product to create horizontal bars for the Gantt chart
for i, product in enumerate(df['Product']):
    # Create a horizontal bar for each product's campaign duration
    plt.barh(product, df['Duration'][i], left=df['StartMonth'][i], color='pink')
plt.title('Gantt Chart of Sales Campaigns')
plt.xlabel('Month Number')
plt.ylabel('Products')
plt.grid(True)
plt.xticks(range(1, 13))
plt.show()

```
